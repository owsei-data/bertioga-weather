```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# mapeia datasets carregados nas variaveis
datasets = {
    2019: data1, 2020: data2, 2021: data3,
    2022: data4, 2023: data5, 2024: data6, 2025: data7
}

# lista para guardar resultados
auditoria = []

# itera sobre cada ano
for ano, df_orig in datasets.items():
    df = df_orig.copy()

    # converte e arruma data
    if 'Data' in df.columns:
        df['DATA_REF'] = pd.to_datetime(df['Data'], errors='coerce')
    elif 'DATETIME_UTC' in df.columns:
        df['DATA_REF'] = pd.to_datetime(df['DATETIME_UTC']).dt.date
        df['DATA_REF'] = pd.to_datetime(df['DATA_REF'])
    else:
        continue

    # extrai ano e mes
    df['Ano'] = df['DATA_REF'].dt.year
    df['Mes'] = df['DATA_REF'].dt.month

    # define colunas alvo
    col_chuva = 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'
    col_vento = 'VENTO, RAJADA MAXIMA (m/s)'

    # verifica se colunas existem
    if col_chuva not in df.columns or col_vento not in df.columns:
         continue

    # agrupa por mes contando registros validos
    resumo = df.groupby(['Ano', 'Mes']).agg(
        Chuva_Valid=(col_chuva, 'count'),
        Vento_Valid=(col_vento, 'count'),
        Total_Horas=('DATA_REF', 'count')
    ).reset_index()

    # calcula porcentagem
    resumo['Pct_Chuva'] = (resumo['Chuva_Valid'] / resumo['Total_Horas']) * 100
    resumo['Pct_Vento'] = (resumo['Vento_Valid'] / resumo['Total_Horas']) * 100

    auditoria.append(resumo)

# gera visualizacao se houver dados
if auditoria:
    df_auditoria = pd.concat(auditoria)

    # heatmap chuva
    plt.figure(figsize=(12, 6))
    heatmap_chuva = df_auditoria.pivot(index='Ano', columns='Mes', values='Pct_Chuva')
    sns.heatmap(heatmap_chuva, annot=True, fmt=".0f", cmap="RdYlGn", vmin=0, vmax=100, cbar_kws={'label': '% Validos'})
    plt.title("Auditoria: CHUVA")
    plt.show()

    # heatmap vento
    plt.figure(figsize=(12, 6))
    heatmap_vento = df_auditoria.pivot(index='Ano', columns='Mes', values='Pct_Vento')
    sns.heatmap(heatmap_vento, annot=True, fmt=".0f", cmap="RdYlGn", vmin=0, vmax=100, cbar_kws={'label': '% Validos'})
    plt.title("Auditoria: VENTO")
    plt.show()

    # relatorio critico
    print("Meses Criticos (Chuva < 70%):")
    criticos = df_auditoria[df_auditoria['Pct_Chuva'] < 70]
    if not criticos.empty:
        print(criticos[['Ano', 'Mes', 'Pct_Chuva']].to_string(index=False))
```

```python
import pandas as pd
import numpy as np

# mapeia os datasets carregados nas variaveis
datasets = {
    2019: data1, 2020: data2, 2021: data3,
    2022: data4, 2023: data5, 2024: data6, 2025: data7
}

# define limite de aceitacao (70%)
LIMITE_ACEITACAO = 70.0

# lista para guardar os agregados
agregados = []

# itera sobre cada ano
for ano, df_orig in datasets.items():
    df = df_orig.copy()

    # converte e arruma data
    if 'Data' in df.columns:
        df['DATA_REF'] = pd.to_datetime(df['Data'], errors='coerce')
    elif 'DATETIME_UTC' in df.columns:
        df['DATA_REF'] = pd.to_datetime(df['DATETIME_UTC']).dt.date
        df['DATA_REF'] = pd.to_datetime(df['DATA_REF'])

    # extrai o mes
    df['Mes'] = df['DATA_REF'].dt.month

    # define colunas alvo
    col_chuva = 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'
    col_vento = 'VENTO, RAJADA MAXIMA (m/s)'

    # agrupa por mes calculando estatisticas e contagem
    mensal = df.groupby('Mes').agg(
        Chuva_Total=(col_chuva, 'sum'),
        Chuva_Max_Dia=(col_chuva, 'max'),
        Vento_Max_Mes=(col_vento, 'max'),
        Vento_Media=(col_vento, 'mean'),
        Valid_Chuva=(col_chuva, 'count'),
        Valid_Vento=(col_vento, 'count'),
        Total_Horas=('DATA_REF', 'count')
    ).reset_index()

    mensal['Ano'] = ano

    # calcula % de dados validos
    mensal['Pct_Chuva'] = (mensal['Valid_Chuva'] / mensal['Total_Horas']) * 100
    mensal['Pct_Vento'] = (mensal['Valid_Vento'] / mensal['Total_Horas']) * 100

    # aplica regra de corte (< 70% vira NaN)
    mask_chuva = mensal['Pct_Chuva'] < LIMITE_ACEITACAO
    mensal.loc[mask_chuva, ['Chuva_Total', 'Chuva_Max_Dia']] = np.nan

    mask_vento = mensal['Pct_Vento'] < LIMITE_ACEITACAO
    mensal.loc[mask_vento, ['Vento_Max_Mes', 'Vento_Media']] = np.nan

    agregados.append(mensal)

# concatena tudo num dataframe so
df_clima_final = pd.concat(agregados).sort_values(by=['Ano', 'Mes'])

# seleciona e ordena colunas finais
cols = ['Ano', 'Mes', 'Chuva_Total', 'Chuva_Max_Dia', 'Vento_Max_Mes', 'Vento_Media', 'Pct_Chuva', 'Pct_Vento']
df_clima_final = df_clima_final[cols]

# salva o csv tratado
df_clima_final.to_csv('weather-bertioga-2019-2025.csv', index=False)

# mostra as primeiras linhas
df_clima_final.head(15)
```
